```python
from google.colab import files
import cv2
from ultralytics import YOLO

# Step 1: Upload video
uploaded = files.upload()
video_path = list(uploaded.keys())[0]  # Get the uploaded file name

print("ðŸ“¹ Uploaded video file:", video_path)

# Step 2: Load YOLO model
model = YOLO("yolov8n.pt")  # or 'yolov8s.pt' for better accuracy

# Step 3: Open video
cap = cv2.VideoCapture(video_path)

# Step 4: Get video properties
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Step 5: Output writer
output_path = "output_detected.mp4"
out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

# Step 6: Process frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)[0]
    annotated_frame = results.plot()

    # Count number of objects in the frame
    object_count = len(results.boxes)

    # Add object count text
    cv2.putText(annotated_frame, f'Objects in frame: {object_count}', (10, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 0, 255), 3)

    out.write(annotated_frame)

# Step 7: Release everything
cap.release()
out.release()

print("âœ… Detection done. Video saved as:", output_path)

# Step 8: Download output
files.download(output_path)
```
